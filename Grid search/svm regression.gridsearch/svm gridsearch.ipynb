```python
#importing the Libraies
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Reading the Dataset
dataset = pd.read_csv('insurance_pre.csv')
dataset
```

```python
dataset=pd.get_dummies(dataset,drop_first=True)
dataset
```

```python
independent=dataset[['age', 'bmi', 'children','sex_male', 'smoker_yes']]
dependent=dataset['charges']
independent
dependent
```

```python
#split into training set and test
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(independent, dependent, test_size = 1, random_state = 0)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()   #StandardScaler converts data into “how far from average” instead of raw values.
X_train = sc.fit_transform(X_train) # age is mostly 40 plus always while bmi is always below that so the model will become bisaed towards the age as it has large value so to give equal importance to it we need to scale all the values 
X_test = sc.transform(X_test)   #converting all the input test and train to scalar value by formula scaled value=x−mean/standard deviation
```

```python
from sklearn.model_selection import GridSearchCV            #importing the gridsearch module for gridsearch
from sklearn.svm import SVR                                 #importing the svr module for svm regression
param_grid = {'kernel':['rbf','poly','sigmoid','linear'],
'C':[10,100,1000,2000,3000],'gamma':['auto','scale']}       # giving all the input parameter to the gridsearch to find the best parameter
grid = GridSearchCV(SVR(), param_grid, refit = True, verbose = 3,n_jobs=-1)   #creating a variable called 'grid' and giving the svm model,all the input parameters ,refit=true to make 0 as no and 1 as yes, into the gridsearch function
grid.fit(X_train, y_train) #ccreating the model by fit function and giving thr input training set
```

```python
re=grid.cv_results_   #storing all the results of the gridsearch intp a variable 're'
print("The R_score value for best parameter {}:".format(grid.best_params_))  # printing the parameter of the best model 
```

```python
table=pd.DataFrame.from_dict(re) 
table  #printing all the result thats stored in the variable 're' 
```
